```python
#In this notebook, we perform SVM an logit model fitting and compare their decision boundaries.

import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import make_blobs


# we use the following simulated data; note that by changing random_state,  you would obtain a different seed
from sklearn.datasets import make_classification
X, y = make_classification(10, 2, 2, 0,random_state=12)



```

```python
import os
#os.getcwd()
```

```python
plt.figure(figsize=[8,6])

plt.scatter(X[:,0],X[:,1],c=y,s=300)
```

```python
# fitting an SVM model with an exponential kernel
#It would be better to scale the data which we ignore doing here.
clf = svm.SVC(kernel='linear', C=100,gamma='scale')
clf.fit(X, y)
#The following command is not available in non-linear kernel
#clf.coef_

#Prediction on the training set
#clf.predict(X)
clf.score(X,y)
```

```python
#Now we perform a logistic regression and compare the result with the SVM
from sklearn.linear_model import LogisticRegression
clf_logit = LogisticRegression(random_state=6,solver='lbfgs').fit(X, y)
```

```python
#The following code graphs the decision boundaries of linear SVM and logistic regression
#Pay attention to the big margin by svm
plt.figure(figsize=[8,6])
plt.scatter(X[:, 0], X[:, 1], c=y, s=30)

# plot the decision function
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# create grid to evaluate model
xx = np.linspace(xlim[0], xlim[1], 50)
yy = np.linspace(ylim[0], ylim[1], 50)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)

# plot decision boundary and margins

ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.9,
           linestyles=['--', '-', '--'])

# plot support vectors
#ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=50,
#           linewidth=1, facecolors='none', edgecolors='k')

grid=np.vstack([XX.ravel(),YY.ravel()]).T
probs = clf_logit.predict_proba(grid)[:, 1].reshape(XX.shape)

ax.contour(XX, YY, probs, levels=[.5], linestyles=['--'],cmap='Reds', vmin=0, vmax=0.7)
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')


plt.show()
```

```python
#We repeat the above experiment but this time using more complicated dataset and non-linear kernel
```

```python
from sklearn.datasets import make_classification
X, y = make_classification(100, 2, 2, 0,random_state=34)

# fitting an SVM model with an exponential kernel
clf = svm.SVC(kernel='rbf', C=100,gamma='scale')
clf.fit(X, y)

#fitting logistic regression
from sklearn.linear_model import LogisticRegression
clf_logit = LogisticRegression(random_state=6,solver='lbfgs').fit(X, y)


plt.figure(figsize=[8,5])
plt.scatter(X[:, 0], X[:, 1], c=y, s=100)

# plot the decision function
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

xx = np.linspace(xlim[0], xlim[1], 50)
yy = np.linspace(ylim[0], ylim[1], 50)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z_svm = clf.decision_function(xy).reshape(XX.shape)

# plot decision boundary and margins

ax.contour(XX, YY, Z_svm, colors='k', levels=[-1, 0, 1], alpha=0.9,
           linestyles=['--', '-', '--'])

grid=np.vstack([XX.ravel(),YY.ravel()]).T
probs = clf_logit.predict_proba(grid)[:, 1].reshape(XX.shape)

ax.contour(XX, YY, probs, levels=[.5], linestyles=['--'],cmap='Reds', vmin=0, vmax=0.7)
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')


plt.show()
```

```python
#Plotting the decision function of SVM which is used in the previous block
from mpl_toolkits.mplot3d import Axes3D  


fig = plt.figure(figsize=[15,15])
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(XX, YY, Z_svm)

ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Decision Function Values')
ax.set_title('Decision Function of SVM: in the case of Gaussian Kernel')

ax.view_init(elev=10., azim=130)

plt.show()
```
